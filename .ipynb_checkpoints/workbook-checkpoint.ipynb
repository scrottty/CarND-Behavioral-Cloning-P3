## Data Augmentation

```python
centerAngles = 0.015
lowerBound = 0.2
lowerBound2 = 0.3
upperBound = 0.98
```

```python
def mimicLeftnRight(angles):
    newAngles = []
    for angle in angles:
        newAngles.append(angle + 0.01)
        newAngles.append(angle - 0.01)
    newAngles = np.append(angles, np.array(newAngles), axis=0)
    return newAngles
```

```python
# Remove zero filter examples
# delRate = 0.98
def filterSamples(labels, threshold, delRate):
    zero_idx = np.where(abs(labels)<=threshold)
    zero_idx = zero_idx[0]
    sizeDel = int(len(zero_idx) * delRate) # select only 10% of the samples
    return np.random.choice(zero_idx, sizeDel, replace=False)
```

```python
def flipImages(images, angles, DoImages=True):
    newImages = []
    newAngles = []
    for image, angle in zip(images, angles):
        if lowerBound2 < abs(angle) < upperBound:
            if DoImages:
                newImages.append(cv2.flip(image, 1))
            newAngles.append(angle * -1)
            
    if len(newAngles) > 0: # To protect against no image transformation
        return np.append(images,np.array(newImages),axis=0), \
                np.append(angles,np.array(newAngles),axis=0)
    else:
        return images, angles
    
```

```python
 # 25º if in the middle of the image therefore 0.15616º for each image shift
# so 0.006 (normalised)
ANGLE_SHIFT = 0.006
def imageShift(images, angles, maxShift=10, DoImages=True):
    shiftedImages = []
    newAngles = []
    for image, angle in zip(images, angles):
        if lowerBound2 < abs(angle) < upperBound:
#             for i in range(0,2):
            shiftX = np.random.uniform(-maxShift, maxShift)
            shiftY = np.random.uniform(-maxShift, maxShift)
            if DoImages:
                M = np.float32([[1,0,shiftX],[0,1,shiftY]])
                image = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))
                shiftedImages.append(image)

            angle = angle + ANGLE_SHIFT * shiftX
            newAngles.append(angle)
    
    if len(newAngles) > 0:
        return np.append(images,np.array(shiftedImages),axis=0),\
                np.append(angles,np.array(newAngles),axis=0)
    else:
        return images, angles
```

```python
def brightnessAugmentation(images, angles, DoImages=True):
    newImages = []
    newAngles = []
    for image, angle in zip(images, angles):
#         if lowerBound < abs(angle) < upperBound:
        if DoImages:
            newImage = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
            brightnessAdjustment = np.random.uniform(0.1,1.1) #+ 0.1 #stop 0
            newImage[:,:,2] = newImage[:,:,2] * brightnessAdjustment
            np.clip(newImage[:,:,2], 0, 255)
            newImages.append(cv2.cvtColor(newImage, cv2.COLOR_HSV2BGR))
        newAngles.append(angle)
    
    if len(newAngles) > 0:    
        return np.array(newImages), np.array(newAngles)
#         return np.append(images, np.array(newImages), axis=0), \
#                 np.append(angles, np.array(newAngles), axis=0)
    else:
        return images, angles
```

```python
# Get labels from dataset
labels = getLabels(samples)
# labels, _ = train_test_split(labels, test_size=0.2)

# filter 98% of the 0.0 samples, 60 of samples < 0.1
delIdx = filterSamples(labels, threshold=centerAngles, delRate=0.98)
labels = np.delete(labels, delIdx, axis=0)
delIdx = filterSamples(labels, threshold=lowerBound, delRate=0.60)
labels = np.delete(labels, delIdx, axis=0)

# Mimic loading the left and right images
labels = mimicLeftnRight(labels)

# Perfrom images flip and shift to samples needing augmentation
_, labels = brightnessAugmentation(labels, labels, DoImages=False)
_, labels = randomShadow(labels, labels, DoImages=False)
_, labels = imageShift(labels, labels, DoImages=False)
_, labels = flipImages(labels, labels, DoImages=False)

num_samples = len(labels)
print('Numple of Data Samples: {:.0f}'.format(num_samples))

# Histogram of values
uniqueValues, uniqueCounts = np.unique(labels, return_counts=True)
print('No. Unique Values: {}'.format(len(uniqueValues)))
print('Most Common Number: {}'.format(uniqueValues[np.argmax(uniqueCounts)]))

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.hist(labels, bins=100)#len(uniqueValues))
plt.box('off')
plt.grid('on')
plt.subplot(1,2,2)
plt.plot(labels)
plt.box('off')
plt.grid('on')
plt.show()
```

## Pre Processing

```python
carWidth = 1.0
recoverydist = 4.0

def loadImages(sample):
    image_center = cv2.imread(sample[0])
    image_left = cv2.imread(sample[1].replace(" ",""))
    image_right = cv2.imread(sample[2].replace(" ",""))
    angle_center = float(sample[3])
    angle_left = angle_center + (carWidth/recoverydist) / 25
    angle_right = angle_center - (carWidth/recoverydist) / 25
    
    return [image_center, image_left, image_right], [angle_center, angle_left, angle_right]
```

```python
# Resize Images for model
def resizeImages(images):
#     images = [cv2.resize(image,(64,64), interpolation=cv2.INTER_AREA) for image in images]
    images = [cv2.resize(image,(200,66), interpolation=cv2.INTER_AREA) for image in images]
    return images
```

```python
def ChangeColourSpace(images):
    images = [cv2.cvtColor(image, cv2.COLOR_BGR2YUV) for image in images]
    return images
```

```python
horizon = 60
bonnet = 140
def cropImages(images, maxShift=10):
    
#     croppedImages = []
#     for image in images:
#         shift = np.random.randint(-maxShift, maxShift)
#         image = image[horizon:bonnet,:]
#         croppedImages.append(image)
#         croppedImages.append(cv2.resize(image, (64,64), cv2.INTER_AREA))
#         croppedImages.append(image[horizon+shift:bonnet+shift,:])
    croppedImages = [image[horizon:bonnet,:] for image in images]
    return croppedImages
```

```python
def generator(samples, batch_size=32):
    while 1:
        samples = shuffle(samples)
#         samplesRemoved = samples
        
        # This augmentation is done here to make it random for each epoch
        angles = getLabels(samples)
        delIdx = filterSamples(np.array(angles), threshold=0.015, delRate=0.98)
        samplesRemoved = np.delete(np.array(samples), delIdx, axis=0)
        
        angles = getLabels(samplesRemoved)
        delIdx = filterSamples(np.array(angles), threshold=0.1, delRate=0.60)
        samplesRemoved = np.delete(np.array(samplesRemoved), delIdx, axis=0)
        
        num_samples = len(samplesRemoved)
        
        for offset in range(0, num_samples, batch_size):
            batch_samples = samplesRemoved[offset:offset+batch_size]
            images = []
            angles = []
            for batch_sample in batch_samples:
                allImages, allAngles = loadImages(batch_sample)
                images.extend(allImages)
                angles.extend(allAngles)
#                 image = cv2.imread(batch_sample[0].replace(" ",""))
#                 images.append(image)
#                 angle = float(batch_sample[1])
#                 angles.append(angle)
            
            # Augmentations
            images, angles = brightnessAugmentation(images, angles)
            images, angles = randomShadow(images, angles)
            images, angles = flipImages(images, angles)
            images, angles = imageShift(images, angles)
#             images = brightnessAugmentation(np.array(images))
            
#             images, angles = imageShift(images, angles)
            
            # Pre Processing
            images = cropImages(np.array(images))
            images = resizeImages(images)
            images = ChangeColourSpace(images)
            X_train = np.array(images)
            y_train = np.array(angles)
            yield shuffle(X_train, y_train)

def generator_valid(samples, batch_size=32):
    while 1:
        samples = shuffle(samples)
        num_samples = len(samples)
        for offset in range(0, num_samples, batch_size):
            batch_samples = samples[offset:offset+batch_size]
            images = []
            angles = []
            for batch_sample in batch_samples:
#                 allImages, allAngles = loadImages(batch_sample)
#                 images.extend(allImages)
#                 angles.extend(allAngles)
                image = cv2.imread(batch_sample[0].replace(" ",""))
                images.append(image)
                angle = float(batch_sample[3])
                angles.append(angle)
            
            images = cropImages(np.array(images))
            images = resizeImages(images)
            images = ChangeColourSpace(images)
            X_valid = np.array(images)
            y_valid = np.array(angles)
            yield shuffle(X_valid, y_valid)


train_samples, valid_samples = train_test_split(samples, test_size=0.2)
train_generator = generator(train_samples, batch_size=32)
valid_generator = generator_valid(valid_samples, batch_size=32)
```

## Models

```python
# Model Selection, Hyperparameters
dropoutRate = 0.5
num_epochs = 20
# num_samples = num_samples * 3 #center, left and right images
num_samples_train = num_samples #int(num_samples*0.8) #0.8 train.valid split, 3 for all cameras
reg = l2(0.001)

print('Number of samples to train on: {}'.format(num_samples_train))
```

### NVIDIA

```python
model = Sequential()
# model.add(Lambda(lambda x: (x/255.0)-0.5, input_shape=(64,64,3)))
model.add(Lambda(lambda x: (x/255.0)-0.5, input_shape=(66,200,3)))
# model.add(Lambda(lambda x: imageStandardise(x), input_shape=(76,320,3)))
model.add(Convolution2D(24,5,5, subsample=(2,2), init = 'he_normal', 
                        activation='elu'))#, W_regularizer = reg))#, input_shape=(76,320,3)))
# model.add(BatchNormalization())
# model.add(Dropout(0.5))
model.add(Convolution2D(36,5,5, subsample=(2,2), init = 'he_normal', 
                        activation='elu'))#, W_regularizer = reg))
# model.add(BatchNormalization())
# model.add(Dropout(0.5))
model.add(Convolution2D(48,5,5, subsample=(2,2), init = 'he_normal',
                        activation='elu'))#, W_regularizer = reg))
# model.add(BatchNormalization())
# model.add(Dropout(0.5))
model.add(Convolution2D(64,3,3, subsample=(1,1), init = 'he_normal',
                        activation='elu'))#, W_regularizer = reg))
# model.add(BatchNormalization())
# model.add(Dropout(0.5))
model.add(Convolution2D(64,3,3, subsample=(1,1), init = 'he_normal',
                        activation='elu'))#, W_regularizer = reg))
# model.add(BatchNormalization())
model.add(Flatten())
model.add(Dropout(0.5))
# model.add(Dense(1164, init = 'he_normal', activation='elu'))
# model.add(Dropout(0.2))
model.add(Dense(100, init = 'he_normal', activation='elu'))#, W_regularizer = reg))
# model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(50, init = 'he_normal', activation='elu'))#, W_regularizer = reg))
# model.add(Dropout(0.5))
model.add(Dense(10, init = 'he_normal', activation='elu'))#, W_regularizer = reg))
# model.add(BatchNormalization())
# model.add(Dropout(0.5))
model.add(Dense(1))

# Compile model
model.compile(loss='mse', optimizer='adam')

# Set up training
filepath = './model-NVIDIA-{epoch:02d}-{val_loss:.5f}.h5'
checkpoint = ModelCheckpoint(filepath, monitor='val_loss', verbose=1, save_best_only=True )
stopper = EarlyStopping(monitor='val_loss', min_delta=.001, patience=5)
callbacks = [checkpoint, stopper, TQDMNotebookCallback(leave_inner=True)] #

history = model.fit_generator(train_generator, samples_per_epoch=num_samples_train, #x2 for birghtness
                   validation_data=valid_generator,
                   nb_val_samples=len(valid_samples), nb_epoch=num_epochs, verbose=0, 
                    callbacks=callbacks);

print('Done! Im Out...')
```

### LeNet

```python
# Resize images before passing to LeNet
model = Sequential()
model.add(Lambda(lambda x: (x/255.0)-0.5, input_shape=(160,320,3)))
model.add(Convolution2D(6,5,5))
model.add(BatchNormalization())
model.add(Activation('relu'))

model.add(MaxPooling2D())

model.add(Convolution2D(16,5,5))
model.add(BatchNormalization())
model.add(Activation('relu'))

model.add(MaxPooling2D())

model.add(Flatten())

model.add(Dense(120))
model.add(BatchNormalization())
model.add(Activation('relu'))
# model.add(Dropout(0.5))

model.add(Dense(84))
model.add(BatchNormalization())
model.add(Activation('relu'))
# model.add(Dropout(0.5))

model.add(Dense(1))

# Compile model
model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])

# Set up training
filepath = './model-LeNet-{epoch:02d}-{val_acc:.4f}.h5'
checkpoint = ModelCheckpoint(filepath, monitor='val_acc', verbose=1, save_best_only=True )
stopper = EarlyStopping(monitor='val_loss', min_delta=.001, patience=2)

model.fit_generator(train_generator, samples_per_epoch=len(train_samples),
                   validation_data=valid_generator,
                   nb_val_samples=len(valid_samples), nb_epoch=50,
                    verbose=2, callbacks=[checkpoint, stopper]);

print('Done! Im Out...')
```
